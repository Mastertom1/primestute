```python
# import the required libraries
import pandas as pd
import seaborn as sns
sns.set(style="darkgrid")
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("popular-languages.csv")
df.head()
```

```python
cols = df.columns
df.drop(columns=cols[0],inplace=True) # remove the column Unnamed:0
df.head()
```

```python
# remove the special character and the commas
col = ["Average Annual Salary"]

# pass them to df.replace(), specifying each char and it's replacement:
df[col] = df[col].replace({'\$': '', ',': ''}, regex=True)
df[col] = df[col].astype(int) 
df.head()
```

```python
df_sorted = df.sort_values("Average Annual Salary", ascending = False)
ax = sns.barplot(data=df_sorted, x="Language", y="Average Annual Salary")
ax.set(title= "Popular Languages")
```
